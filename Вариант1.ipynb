У вас есть размеченный файл  словаря - Dictionary.xml, в словаре содержатся записи двух типов: леммы (lemma) и словоформы (inflected). Записи выглядят слующим образом
```
<entry>
          <!-- лемма -->
          <form type="lemma"><orth>go</orth></form> 
          <!-- часть речи -->
          <gramGrp><gram type="partOfSpeech">verb</gram></gramGrp>
          <!-- список словоформ, если слово изменяемое -->
          <form type="inflected">
            <orth>went</orth>
          </form>
          <form type="inflected">
            <orth>gone</orth>
          </form>
</entry>
```
Вам нужно 

1. Создать класс ***Dict*** в котором будут следующие методы 
    + ***parse_dict(self, path: str)*** - принимает на вход путь к файлу словаря, парсит файл, создает список объектов Lemma и Inflected, сохраняет его в атрибут 
    + ***get_sorted_objects(self)*** - возвращает список объектов (именно объектов, а не строк) словаря, отсортированных по алфавиту (все вместе: и леммы, и словоформы)
    + ***find(self, word: str)*** -  принимает на вход искомое слово (учитывается только полное соответствие, а не вхождение в качестве подстроки) возвращает объект, если слово есть в словаре или None, если его нет в словаре
    

2. Записи в словаре должны быть представлены в виде объектов классов Lemma и Inflected
   + у объектов обоих классов должен быть атрибут ***text*** в котором содержится собственно текст леммы или словоформы, сортировка объектов по алфавиту производится по этому атрибуту 
   + у объектов обоих классов должен быть атрибут ***pos***, в котором хранится название части речи
   + у объектов класса ***Lemma*** должен быть метод ***get_wordforms(self)***, который вернет все принадлежащие лемме словоформы (объекты класса Inflected) в виде генератора

**Полезная информация:**
+ Старайтесь не дублировать код! Толковое применение пройденных вещей (наследование, полиморфизм и т.д.) приветствуется. За некачественный и неоптимальный код будем снижать. 
+ xml очень похож на html, парсить удобно с помощью lxml или BeautifulSoup 
+ Чтобы отсортировать объекты Lemma/Inflected в алфавитном порядке их должно быть можно сравнивать между собой (должны работать операторы (<, > и = )
+ Строки в питоне можно сравнивать между собой обычными оперторами сравнения, меньше будет та, которая в алфавитном порядке стояла бы раньше (но будьте внимательны, символы в верхнем регистре идут раньше символов в нижнем, так что лучше привести все к одному регистру при сравнении)
```
'aaa' < 'aab' 
>>>> True
'bbb' < 'aaa'
>>>> False
'a' < 'B'
>>>> False
'A' < 'B'
>>>> True
```
